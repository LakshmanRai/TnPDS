```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
df = pd.read_csv('/content/TNPDS.csv')

df.dropna(inplace=True)
df.drop(41, inplace=True)
df.drop(42, inplace=True)

df['DISTRICT NAME']=df['DISTRICT NAME'].astype('string')
df['TOTAL NUMBER OF TALUKS']=df['TOTAL NUMBER OF TALUKS'].astype('int')
df['TOTAL NUMBER OF FPS']=df['TOTAL NUMBER OF FPS'].astype('int')
df['TOTAL NUMBER OF CARDS']=df['TOTAL NUMBER OF CARDS'].astype('int')
df['TOTAL NUMBER OF BENEFICIARIES']=df['TOTAL NUMBER OF BENEFICIARIES'].astype('int')
df['TOTAL NUMBER OF AADHAAR REGD.']=df['TOTAL NUMBER OF AADHAAR REGD.'].astype('int')
df['TOTAL NUMBER OF MOBILE REGD.']=df['TOTAL NUMBER OF MOBILE REGD.'].astype('int')

df.to_csv('/content/TNPDS.csv')
```

```python
df
```

```python
mp.figure(figsize=(15,3))
mp.bar(df['DISTRICT NAME'], df['TOTAL NUMBER OF TALUKS'])
mp.xticks(rotation=90)
mp.title('Taluks district wise')
mp.xlabel('Districts')
mp.ylabel('No. of. Taluks')
mp.show()
```

```python
mp.figure(figsize=(15,3))
mp.grid(color='g', linestyle='dotted')
sns.boxplot(data=df, x=df['DISTRICT NAME'], y=df['TOTAL NUMBER OF BENEFICIARIES'])
mp.xticks(rotation=90)
mp.title('Beneficiaries')
mp.xlabel('Districts')
mp.ylabel('No. of. Beneficiaries')
mp.show()
```

```python
mp.figure(figsize=(15,3))
mp.grid(color='y',linestyle='dashed')
mp.title('Cards')
mp.xlabel('District')
mp.ylabel('No. of cards')
sns.pointplot(df, x=df['DISTRICT NAME'], y=df['TOTAL NUMBER OF CARDS'])
mp.xticks(rotation=90)
mp.show()
```

```python
mp.figure(figsize=(12, 5))
sns.scatterplot(df, x=df['TOTAL NUMBER OF AADHAAR REGD.'], y=df['TOTAL NUMBER OF MOBILE REGD.'])
mp.xlabel('No. of Aadhaar Registrations')
mp.ylabel('No. of Mobile Registrations')
mp.title('Aadhaar Registrations vs Mobile Registrations')
mp.show()
```

```python
mp.figure(figsize=(12, 3))
sns.swarmplot(df, x=df['TOTAL NUMBER OF CARDS'], y=df['TOTAL NUMBER OF BENEFICIARIES'])
mp.title('Beneficiaries vs. Cards')
mp.xticks(rotation=90)
mp.show()
```
